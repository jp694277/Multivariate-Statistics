#### 想一想你目前最有兴趣研究的一个网络是什么，并回答以下问题: 

我想研究的网络关系是**社会网络爱情关系**。在社会网络中，一个人的网络邻居在不同的社会情境中具有相当重要的作用，如社会支持、工作机会。网络结构中包括了家庭成员、同事、长期朋友、熟人，甚至包括潜在的恋人或配偶，而恋人或配偶往往是给予人们最大心灵陪伴的角色。此外，与其他亲人、朋友的关系相比，爱情关系构成了个体每天实践和社交媒体使用，因此决定研究社会网络爱情关系，以下的问题设计参考Lars Backstrom和Jon Kleinberg于2013年在Eprint Arxiv上发表的《[Romantic Partnerships and the Dispersion of Social Ties: A Network Analysis of Relationship Status on Facebook](https://arxiv.org/pdf/1310.6753.pdf)》：

- **这个网络的节点是什么?它们之间的互动关系是什么?**     
这个网络的节点是“个体”，也就是社会网络上的每个人，而人与人间的互动关系则为“联系”。   
   
   
    
- **你觉得该怎样来收集相关数据，或者有什么数据源可以用来开展这项研究?**   
现在是互联网时代，又恰逢新冠疫情在现实世界中肆虐，因此使用网络来社交成为非常主流的方式，社会网络逐渐转为以网上互动呈现。关于这项研究，可以搜集微博、facebook、twitter...等网络社交平台的数据，并凭藉用户标示的感情状态如：“已婚”、“恋爱中”、“单身”...等，来辨识相互间的关系。可以通过爬虫或是由数据库获取数据的方式，在个人网络资料中筛选出标示配偶或恋人关系的成年人，这些人的连接的社会网络包含了众多与他人的关系。   
   
   
   
- **这个网络的规模有多大?即怎样划定你的研究对象的范围，大致由多少个节点组成?**      
根据Lars Backstrom和Jon Kleinberg的研究内容，该研究采用了130万名个体用户的数据，而这些个体用户连接的社会网络包含了86亿个与他人的关系。根据这个研究结果，我估计研究对象的数据量乘上6615，即可得出整个社会网络的研究范围。   
下图展示了该研究中某个用户的网络结构，图的上侧和右上侧有两个密度较高，相互独立的网络，其余网络则较为稀疏，可见个体的网络结构是异质性的。因此可结论出跨群的个体作为连接桥，对于理解网络中的关系非常重要。   


    接着则是来定义每一位个体用户大致由多少节点组成，根据该研究的方法，可以设计出一个由**11个节点组成的网络**，再依序扩展出整个社会网络。在社会网络爱情关系的网络设计中，需要舍弃较常使用的嵌入式网络（Network embeddedness），转而使用**分散式网络（Network Dispersion）**。事实上一般而论，在生活中多数人都拥有不少朋友群，这些群在互动中形成社会焦点（Social Foci），如同事群、同学群。而在群里，人们若彼此相互认识，即使没有强关系，这些群也包含了较高的嵌入性。然而，一个人的伴侣或其他关系密切的朋友可能只有低度嵌入性，在这个网络中，个体间通过不同的几个焦点相互联系，但没有一个高度重叠的焦点，例如一个成年男性，他妻子有同事、亲戚、同学，这些人可以通过这个妻子连接起来，但他们却相互不认识，因此选定分散式网络。   
    下图显示了用户$u$的社会网络，为单独个体的网络设计方式。了解此图方法如下：若想辨别$u$与$b、c、f、h$间的分散程度，则可以先看崁入性和成对性。若从$u$到$b、c、f$每个连接都有5度的嵌入性，而$u$和$h$只有4度嵌入性；另一方面，$u$和$h$在$c$和$f$、$j$和$k$的连接中是独特的一对，由这两个方面来看则可判别，$u-h$之间的连接分散度高于$u$与$b、c、f$之间的连接。而我们想看的就是连接分散度低的节点。     


- **你对这个网络中的什么现象或问题有兴趣?为什么?**   
在Lars Backstrom和Jon Kleinberg的研究中，甚至可以预测，如果一对恋人或配偶的分散性得分比较低，他们在随后两个月分手的机率会大于50%，并且结婚人士的预测效力更高，尤其是已婚男性。并且确实结论出分散度越高，两个人形成伴侣关系的可能性越大，同性伴侣关系也是如此。   
**“社会网络分散度越高，两个人形成伴侣关系的可能性越大”**，这个现象让我感到十分有兴趣。一般来说，人们可能会直觉性地认为伴侣和个体间的社会焦点（Social Foci）会有较高的重叠焦点，然而实际上却是没有的，反而越低度导致分散度越高，更可能形成伴侣关系。这个现象可以配合社会心理学相关的专家来研究，以寻找出一个合理的解释。   
而就我自身的经验来看，虽然没有已婚的朋友，然而有伴侣的朋友们，若是感情叫稳固且交往时间较长久，其伴侣确实和该个体的社会群体分散度较高。以往我也和不少人谈论分享过对于伴侣的选择议题，会发现不少男性在追求伴侣的阶段，便会选择自己的社交群体外的个体作为可能发展对象，原因有许多如：不希望这段爱情关系未来或恐破坏原先的社交群体、害怕爱情关系被共同社交群体过度关注、担忧失败的爱情关系被共同社交群体发觉、认为较不熟识的个体相对有吸引力...等。未来若有机会研究相关网络，我会选择和社会心理相关的研究者合作，一解这个神奇的议题。

```python
df = pd.read_csv('/Users/wangziwen/Documents/Graduation/1st/Big Data/Big Data Analysis Base/HW2/samgov.csv')
#df.info() #了解数据结构
```

```python
# 数据预处理
df_words = df.content[0]
for i in range(696):
    df_words = df_words + df.content[i+1]

seg_list = jieba.lcut(df_words, cut_all=False) # 分词（包含停词）

# 处理停留词
def remove_stop_words(file_name,seg_list):
    with open(file_name,'r', encoding="utf8", errors='ignore') as f:
        stop_words = f.readlines()
    stop_words = [stop_word.rstrip() for stop_word in stop_words]
    new_list=[]
    stop_list=[]
    for seg in seg_list:
        if seg not in stop_words:
            new_list.append(seg) # 去除停词的分词
    for seg in seg_list:
        if seg in stop_words:
            stop_list.append(seg) # 只有停词的分词
    return new_list,stop_list
file_name = './stop.txt' # https://raw.githubusercontent.com/tomlinNTUB/Python/master/中文分詞/停用詞.txt
seg_list,stoplist = remove_stop_words(file_name,seg_list) # 分成去除停词的分词＆只有停词的分词

# 统计词频
def count_segment_freq(seg_list):
    seg_df = pd.DataFrame(seg_list,columns=['seg'])
    seg_df['count'] = 1
    sef_freq = seg_df.groupby('seg')['count'].sum().sort_values(ascending=False)
    sef_freq = pd.DataFrame(sef_freq)
    return sef_freq
```

首先将每一个`ID`的文本全部合并成一个list，接着导[停用词.txt](https://raw.githubusercontent.com/tomlinNTUB/Python/master/中文分詞/停用詞.txt)作为停用词的字典，而后以此停用词字典为主，来将分词区分为**去除停词的分词**和**只有停词的分词数**，并分别对此进行词频统计，下面可以看出去除停词的分词数和只有停词的分词数：

```python
# 去除停词的分词词频
sef_freq = count_segment_freq(seg_list)
sef_freq['freq'] = sef_freq['count']/sef_freq['count'].sum()

# 只有停词的分词词频
stop_freq = count_segment_freq(stoplist)
stop_freq['freq'] = stop_freq['count']/stop_freq['count'].sum()

print('去除停词的分词数：',sef_freq['count'].sum())
print('只有停词的分词数：',stop_freq['count'].sum())
```

接着可以看**去除停词的分词词频统计**，列出频率最高的前十名分词，可以看出此文本可能是**“对政府反映意见”**：

```python
# 文本集
l = [df.content[0]]
for i in range(696):
    l.append(df.content[i+1])
# 分词
words_ls = []
for i in range(697):
    seg_list = jieba.lcut(l[i], cut_all=False) # 分词（包含停词）
    seg_list,stoplist = remove_stop_words(file_name,seg_list) # 分成去除停词的分词＆只有停词的分词
    words_ls.append(seg_list)
    
# 构造词典
dictionary = corpora.Dictionary(words_ls)
# 基于词典，使【词】→【稀疏向量】，并将向量放入列表，形成【稀疏向量集】
corpus = [dictionary.doc2bow(words) for words in words_ls]
# lda模型，num_topics设置主题的个数
lda = models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=10)
# 打印所有主题，每个主题显示5个词
for topic in lda.print_topics(num_words=10):
    print(topic)
# 主题推断
#print(lda.inference(corpus))
LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda, corpus, dictionary)
pyLDAvis.display(LDAvis_prepared)
```
